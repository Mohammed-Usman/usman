```python
import pandas as pd
import numpy as np
from keras import models
from keras import layers
from keras.models import Sequential
from keras.layers import Dense
```

```python
np.random.seed(7)
#df = pd.read_csv('data.csv')
#data = np.array(df)
#print(data)

data = np.loadtxt("data.csv",delimiter = ",")
print(data)
len(data)
'''A delimiter is a sequence of one or more characters used to specify the boundary between separate,
independent regions in plain text or other data streams.
An example of a delimiter is the comma character,
which acts as a field delimiter in a sequence of comma-separated values.'''
```

```python
def build_models():
    
    model=Sequential()
    model.add(Dense(12,input_dim = 8,activation ='relu'))
    model.add(Dense(8,activation = 'relu'))
    model.add(Dense(1,activation = 'sigmoid'))
    model.compile(loss = 'binary_crossentropy',optimizer = 'adam' , metrics = ['accuracy'])
    return model
```

```python
k = 4
num_validation_samples = len(data) // k
np.random.shuffle(data)
validation_scores = []
for fold in range(k):
    validation_data = data[num_validation_samples * fold : num_validation_samples * (fold + 1)]
    print(len(validation_data))
    training_data = np.vstack((data[:num_validation_samples * fold] , data[num_validation_samples * (fold + 1):]))
    print(len(training_data))
    
    X = training_data[:,0:8]
    Y = training_data[:,8]
   
    
    model = build_models()
    model.fit(X,Y,epochs = 100,batch_size = 15)
    
    #training_data = data[:]
    # model = get_model()
    #model.train(training_data)
    #model.train(training_data)
    X_Val = validation_data[:,0:8]
    Y_Val = validation_data[:,8]
    a,b=model.evaluate(X_Val,Y_Val)
    print('X_Val: ',a)
    print('Y_Val: ',b)
    print('X_Train: ',len(X))
    print('Y_Train: ',len(Y))
    
#validation_score = np.average(validation_scores)
#model = build_models()
#model.train(data)
#test_score = model.evaluate(test_data)
```
